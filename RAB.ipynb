```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Defina o caminho para salvar o arquivo após o upload
file_path = '/content/V_OCORRENCIA_AMPLA.csv'

# Verifica se o arquivo já existe para evitar upload repetido
if not os.path.exists(file_path):
    # Solicita que o usuário faça o upload do arquivo se ele ainda não foi carregado
    uploaded = files.upload()

    # Salva o arquivo carregado no caminho especificado
    for filename in uploaded.keys():
        os.rename(filename, file_path)
        print(f"Arquivo {filename} carregado e salvo como {file_path}")

# Carrega o arquivo CSV no DataFrame
data = pd.read_csv(file_path, delimiter=';')
print("Dados carregados com sucesso!")

# Exibe as primeiras linhas para confirmar o carregamento
print(data.head())

# Código de análise (exemplo: contagem de 'Classificacao_da_Ocorrencia')
if 'Classificacao_da_Ocorrencia' in data.columns:
    occurrence_type_counts = data['Classificacao_da_Ocorrencia'].value_counts()
    print("Tipos de Ocorrência mais comuns:\n", occurrence_type_counts)

    # Gráfico dos tipos de ocorrência mais comuns
    occurrence_type_counts.plot(kind='bar', title="Most Common Occurrence Types")
    plt.xlabel("Occurrence Classification")
    plt.ylabel("Frequency")
    plt.show()
else:
    print("A coluna 'Classificacao_da_Ocorrencia' não foi encontrada no DataFrame.")

import pandas as pd

# ajustar o `skiprows` após inspecionar as primeiras linhas
data = pd.read_csv('/content/V_OCORRENCIA_AMPLA.csv', delimiter=';', skiprows=1)  # Ajuste `skiprows` se necessário

# Exibir as colunas carregadas no DataFrame
print("Colunas disponíveis no DataFrame:", data.columns)

# Contagem dos tipos de ocorrência na coluna 'Classificacao_da_Ocorrencia'
occurrence_type_counts = data['Classificacao_da_Ocorrencia'].value_counts()
print("Tipos de Ocorrência mais comuns:\n", occurrence_type_counts)

# Gráfico dos tipos de ocorrência mais comuns
import matplotlib.pyplot as plt

# Exibir a contagem de todos os tipos de ocorrência na coluna 'Classificacao_da_Ocorrencia'
occurrence_type_counts = data['Classificacao_da_Ocorrencia'].value_counts()
print("Todos os tipos de ocorrência:\n", occurrence_type_counts)

# Exibir todos os tipos de ocorrência em um gráfico de barras
import matplotlib.pyplot as plt

occurrence_type_counts.plot(kind='bar', title="Todos os Tipos de Ocorrência")
plt.xlabel("Classificação da Ocorrência")
plt.ylabel("Frequência")
plt.show()
# Converter a coluna 'Data_da_Ocorrencia' para o tipo datetime
data['Data_da_Ocorrencia'] = pd.to_datetime(data['Data_da_Ocorrencia'], errors='coerce')

# Exibir a data mais antiga e a mais recente na base
data_inicio = data['Data_da_Ocorrencia'].min()
data_fim = data['Data_da_Ocorrencia'].max()

print(f"Data mais antiga da base: {data_inicio}")
print(f"Data mais recente da base: {data_fim}")

#Pergunta: 1 - Qual é a descrição mais comum para tipos específicos de incidentes?

# Count descriptions in 'Descricao_do_Tipo'
description_counts = data['Descricao_do_Tipo'].value_counts()
print(description_counts)
# Plot the top 10 common descriptions
description_counts.head(10).plot(kind='bar', title="Descrições de incidentes mais comuns")
plt.xlabel("Incident Description")
plt.ylabel("Frequency")
plt.show()

# 2. Perguntas sobre gravidade e danos
Pergunta: Quais são os níveis mais comuns de danos em aeronaves?

# Count levels of damage in 'Danos_a_Aeronave'
damage_counts = data['Danos_a_Aeronave'].value_counts()
print(damage_counts)
# Plot the most common levels of damage
damage_counts.plot(kind='bar', title="Níveis mais comuns de danos em aeronaves")
plt.xlabel("Danos à aeronave")
plt.ylabel("Frequência")
plt.show()

#Pergunta: 2.1 Quais incidentes resultaram em ferimentos fatais?
# Filtrar dados com ferimentos fatais
fatal_injuries = data[(data['Lesoes_Fatais_Tripulantes'] > 0) |
                      (data['Lesoes_Fatais_Passageiros'] > 0) |
                      (data['Lesoes_Fatais_Terceiros'] > 0)]

# Exibir as colunas relevantes para esses incidentes
print(fatal_injuries[['Classificacao_da_Ocorrencia', 'Lesoes_Fatais_Tripulantes',
                      'Lesoes_Fatais_Passageiros', 'Lesoes_Fatais_Terceiros',
                      'Data_da_Ocorrencia', 'Municipio', 'UF']])

# 2.3 Pergunta: Certas operações ou fases de voo têm maior probabilidade de causar danos graves?
# Filter data for severe damage
severe_damage = data[data['Danos_a_Aeronave'] == 'Destruída']
operation_phase_counts = severe_damage['Fase_da_Operacao'].value_counts()
operation_counts = severe_damage['Operacao'].value_counts()

print("Operation Phases with Severe Damage:", operation_phase_counts)
print("Operation Types with Severe Damage:", operation_counts)

# Plot the operation phases with the most severe damage
operation_phase_counts.plot(kind='bar', title="Flight Phases with Most Severe Damage")
plt.xlabel("Flight Phase")
plt.ylabel("Frequency")
plt.show()

# 3.Perguntas sobre localização e sazonalidade
#Pergunta: Em quais regiões e estados ocorrem mais incidentes?
# Count incidents by region and state
region_counts = data['Regiao'].value_counts()
state_counts = data['UF'].value_counts()

print("Incidents by Region:", region_counts)
print("Incidents by State:", state_counts)

# Plot incidents by region
region_counts.plot(kind='bar', title="Incidents by Region")
plt.xlabel("Region")
plt.ylabel("Frequency")
plt.show()

# 3.2 Pergunta: Existe sazonalidade nos incidentes ao longo do ano?
# Convert 'Data_da_Ocorrencia' to datetime and extract the month
data['Data_da_Ocorrencia'] = pd.to_datetime(data['Data_da_Ocorrencia'], errors='coerce')
data['Month'] = data['Data_da_Ocorrencia'].dt.month

# Count incidents by month
month_counts = data['Month'].value_counts().sort_index()
print("Incidents by Month:", month_counts)

# Plot seasonality by month
month_counts.plot(kind='bar', title="Seasonality of Incidents Throughout the Year")
plt.xlabel("Month")
plt.ylabel("Frequency")
plt.show()

# 4. Perguntas sobre as características da aeronave
# Pergunta: Quais tipos e categorias de aeronaves estão mais envolvidas em incidentes?

# Count incidents by model and category of aircraft
model_counts = data['Modelo'].value_counts()
category_counts = data['Categoria_da_Aeronave'].value_counts()

print("Incidents by Aircraft Model:", model_counts.head())
print("Incidents by Aircraft Category:", category_counts)

# Plot incidents by aircraft category
category_counts.plot(kind='bar', title="Most Involved Aircraft Categories in Incidents")
plt.xlabel("Aircraft Category")
plt.ylabel("Frequency")
plt.show()

# 4.1 Pergunta: Existe uma correlação entre o Peso Máximo de Decolagem (PMD) e a gravidade dos incidentes?
# Convert PMD to numeric and analyze correlation with damage levels
data['PMD'] = pd.to_numeric(data['PMD'], errors='coerce')
pmd_damage_correlation = data[['PMD', 'Danos_a_Aeronave']].dropna()

# Plot the correlation between PMD and damage severity
sns.boxplot(x='Danos_a_Aeronave', y='PMD', data=pmd_damage_correlation)
plt.title("Correlation between PMD and Incident Severity")
plt.xlabel("Aircraft Damage")
plt.ylabel("Maximum Takeoff Weight (PMD)")
plt.show()

#5. Perguntas sobre o contexto das operações de voo
#Pergunta: Quais tipos de operação são mais propensos a incidentes graves?
# Count operation types in severe incidents
severe_operations = data[data['Danos_a_Aeronave'] == 'Destruída']
operation_type_counts = severe_operations['Operacao'].value_counts()
print(operation_type_counts)

# Plot the operation types with the most severe incidents
operation_type_counts.plot(kind='bar', title="Operations Prone to Severe Incidents")
plt.xlabel("Operation Type")
plt.ylabel("Frequency")
plt.show()

#5.2 Pergunta: Durante qual fase do voo ocorrem mais incidentes?
# Count flight phases with incidents
flight_phase_counts = data['Fase_da_Operacao'].value_counts()
print(flight_phase_counts)

# Plot flight phases with the most incidents
flight_phase_counts.plot(kind='bar', title="Flight Phases with Most Incidents")
plt.xlabel("Flight Phase")
plt.ylabel("Frequency")
plt.show()

# 6.1. Distribuição de Lesões entre Tripulantes, Passageiros e Terceiros
# Selecionar colunas de lesões
lesoes_cols = ['Lesoes_Fatais_Tripulantes', 'Lesoes_Fatais_Passageiros', 'Lesoes_Fatais_Terceiros',
               'Lesoes_Graves_Tripulantes', 'Lesoes_Graves_Passageiros', 'Lesoes_Graves_Terceiros',
               'Lesoes_Leves_Tripulantes', 'Lesoes_Leves_Passageiros', 'Lesoes_Leves_Terceiros']

# Somar o total de lesões por categoria
lesoes_totais = data[lesoes_cols].sum()

# Gráfico de barras para distribuição de lesões
plt.figure(figsize=(12, 6))
lesoes_totais.plot(kind='bar', color='cornflowerblue')
plt.title("Distribuição de Lesões entre Tripulantes, Passageiros e Terceiros")
plt.xlabel("Tipo de Lesão")
plt.ylabel("Número de Lesões")
plt.xticks(rotation=45)
plt.show()

#6.2. Ocorrências com Lesões Desconhecidas
# Selecionar colunas de lesões desconhecidas
lesoes_desconhecidas_cols = ['Lesoes_Desconhecidas_Tripulantes', 'Lesoes_Desconhecidas_Passageiros', 'Lesoes_Desconhecidas_Terceiros']

# Somar o total de lesões desconhecidas
lesoes_desconhecidas_totais = data[lesoes_desconhecidas_cols].sum()

# Gráfico de barras para lesões desconhecidas
plt.figure(figsize=(10, 5))
lesoes_desconhecidas_totais.plot(kind='bar', color='salmon')
plt.title("Ocorrências com Lesões Desconhecidas")
plt.xlabel("Tipo de Lesão Desconhecida")
plt.ylabel("Número de Lesões")
plt.xticks(rotation=45)
plt.show()

#6.3. Análise Temporal das Ocorrências ao Longo dos Anos
# Extrair o ano da data da ocorrência
data['Ano'] = data['Data_da_Ocorrencia'].dt.year

# Contar o número de ocorrências por ano
ocorrencias_por_ano = data['Ano'].value_counts().sort_index()

# Gráfico de linhas para evolução das ocorrências ao longo dos anos
plt.figure(figsize=(12, 6))
ocorrencias_por_ano.plot(kind='line', marker='o', color='teal')
plt.title("Evolução das Ocorrências ao Longo dos Anos")
plt.xlabel("Ano")
plt.ylabel("Número de Ocorrências")
plt.grid(True)
plt.show()

#6.4. Identificação de Fatores Importantes para a Gravidade dos Incidentes
#Para identificar os fatores mais importantes para a gravidade dos incidentes,
#podemos observar a relação entre o tipo de ocorrência, o nível de dano e a presença de lesões.

# Filtrar colunas relevantes para análise de gravidade
fatores_gravidade = data[['Tipo_de_Ocorrencia', 'Danos_a_Aeronave', 'Lesoes_Fatais_Tripulantes',
                          'Lesoes_Graves_Tripulantes', 'Lesoes_Leves_Tripulantes']].copy()

# Criar uma nova coluna que soma todos os tipos de lesões nos tripulantes
fatores_gravidade['Total_Lesoes_Tripulantes'] = fatores_gravidade[['Lesoes_Fatais_Tripulantes',
                                                                   'Lesoes_Graves_Tripulantes',
                                                                   'Lesoes_Leves_Tripulantes']].sum(axis=1)

# Plot da gravidade dos danos com relação ao tipo de ocorrência
plt.figure(figsize=(12, 8))
sns.boxplot(data=fatores_gravidade, x='Danos_a_Aeronave', y='Total_Lesoes_Tripulantes', hue='Tipo_de_Ocorrencia')
plt.title("Gravidade dos Incidentes por Tipo de Ocorrência e Nível de Dano")
plt.xlabel("Nível de Dano à Aeronave")
plt.ylabel("Total de Lesões em Tripulantes")
plt.legend(title="Tipo de Ocorrência", loc='upper right')
plt.show()

#7. Como as Ocorrências Mudaram ao Longo dos Anos?
# Supondo que a coluna 'Data_da_Ocorrencia' já está no formato datetime
data['Data_da_Ocorrencia'] = pd.to_datetime(data['Data_da_Ocorrencia'], errors='coerce')

# Extrair o ano da data de ocorrência
data['Ano'] = data['Data_da_Ocorrencia'].dt.year

# Contar o número de ocorrências por ano
ocorrencias_por_ano = data['Ano'].value_counts().sort_index()

# Gráfico de linhas para evolução das ocorrências ao longo dos anos
plt.figure(figsize=(12, 6))
ocorrencias_por_ano.plot(kind='line', marker='o', color='teal')
plt.title("Evolução das Ocorrências ao Longo dos Anos")
plt.xlabel("Ano")
plt.ylabel("Número de Ocorrências")
plt.grid(True)
plt.show()

#7.2 2. Quais Fatores São Mais Importantes para a Gravidade dos Incidentes?
#Para essa análise, vamos considerar as variáveis 'Tipo_de_Ocorrencia'​​,
#'Danos_a_Aeronave'e as colunas de lesões ( 'Lesoes_*').
#Vamos usar gráficos de boxplot para explorar a relação entre essas variáveis ​​e a gravidade dos incidentes.

# Filtrar colunas relevantes para análise de gravidade
fatores_gravidade = data[['Tipo_de_Ocorrencia', 'Danos_a_Aeronave', 'Lesoes_Fatais_Tripulantes',
                          'Lesoes_Graves_Tripulantes', 'Lesoes_Leves_Tripulantes']].copy()

# Criar uma nova coluna que soma todos os tipos de lesões nos tripulantes
fatores_gravidade['Total_Lesoes_Tripulantes'] = fatores_gravidade[['Lesoes_Fatais_Tripulantes',
                                                                   'Lesoes_Graves_Tripulantes',
                                                                   'Lesoes_Leves_Tripulantes']].sum(axis=1)

# Gráfico de boxplot para mostrar a relação entre danos, tipo de ocorrência e total de lesões
plt.figure(figsize=(12, 8))
sns.boxplot(data=fatores_gravidade, x='Danos_a_Aeronave', y='Total_Lesoes_Tripulantes', hue='Tipo_de_Ocorrencia', palette='Set3')
plt.title("Gravidade dos Incidentes por Tipo de Ocorrência e Nível de Dano")
plt.xlabel("Nível de Dano à Aeronave")
plt.ylabel("Total de Lesões em Tripulantes")
plt.legend(title="Tipo de Ocorrência", loc='upper right')
plt.show()


```
